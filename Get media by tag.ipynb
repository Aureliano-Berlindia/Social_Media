# 按照Tag爬取media

```python
medias = instagram.get_medias_by_tag('galanz', count=999999)
```

```python
len(medias)
```

```python
cols = ['Date-Time','Year','Month','Day','Time','Like','Comment','Text','Link','Post_id','User_id']
df = pd.DataFrame(columns = cols)
df
```

```python
for i in range(len(medias)):
    #### 总时间-年月日-时分 ####
    post_info = []
    timearray = time.localtime(medias[i].created_time)
    post_info.append(time.strftime("%Y-%m-%d %H:%M:%S", timearray))
    post_info.append(time.strftime("%Y", timearray))
    post_info.append(time.strftime("%m", timearray))
    post_info.append(time.strftime("%d", timearray))
    post_info.append(time.strftime("%H:%M:%S", timearray))       #which fucking timezone?


    post_info.append(medias[i].likes_count)
    post_info.append(medias[i].comments_count)
    
    ### content-like-comment  ###
    post_info.append(medias[i].caption)

    post_info.append(medias[i].link)
    post_info.append(str(medias[i].identifier))
    post_info.append(str(medias[i].owner.identifier))
    x = pd.Series(post_info,index=df.columns)
    df = df.append(x,ignore_index = True)
```

# 爬取发帖者信息

```python
df['User_name'] = None
df['Fans'] = None
df['Follows'] = None
df.head()
```

```python
for i in tqdm(range(len(df))):

    try:
        if i%10==0 and i!=0:
            time.sleep(300)
        elif i%2==0 and i!=0:
            time.sleep(120)
        
        account = instagram.get_account_by_id(df['User_id'][i])
        df['Fans'][i] = account.followed_by_count
        df['Follows'][i] = account.follows_count
        df['User_name'][i] = account.username
    
    except Exception as e:
        error_info = []
        error_info.append(str(e))
        print(i)
```
